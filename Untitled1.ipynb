#### Define function which generates features from image names 

```python
to_tensor = transforms.ToTensor()
transform = transforms.Compose([
        transforms.Resize([224,224]),
        transforms.ToTensor(),
        ])

def get_vector(image_name):
    # 1. Load the image with Pillow library
    img = Image.open(image_name)
    #imshow(np.asarray(img))
    #plt.draw()
    # 2. Create a PyTorch Variable with the transformed image
    t_img = Variable(transform(img).unsqueeze(0))
    return features(t_img).reshape(-1,1)
```

```python
class Network(nn.Module):
    def __init__(self, emb_dim=128):
        super(Network, self).__init__()
        
        self.fc = nn.Sequential(
            nn.Linear(2048, 512),
            nn.Dropout(0.1),
            nn.LeakyReLU(),
            nn.Linear(512, 512),
            nn.LeakyReLU(),
            nn.Linear(512, 128),
            nn.LeakyReLU(),
            nn.Linear(128, 128),
            nn.Dropout(0.1),
            nn.LeakyReLU(),
            nn.Linear(128, emb_dim),
            nn.LeakyReLU(),
        )
        
    def forward(self, x):
        x = self.fc(x)
        return x
```

```python
def distance(a,b):
    a = a.squeeze()
    b = b.squeeze()
    d = torch.dot((a-b).T,(a-b))/(torch.norm(a)*torch.norm(b))
    return d.item()

class TripletLoss(nn.Module):
    def __init__(self, margin=1.0):
        super(TripletLoss, self).__init__()
        self.margin = margin
        
    def calc_euclidean(self, x1, x2):
        return (x1 - x2).pow(2).sum(1)
    
    def forward(self, anchor: torch.Tensor, positive: torch.Tensor, negative: torch.Tensor) -> torch.Tensor:
        distance_positive = self.calc_euclidean(anchor, positive)
        distance_negative = self.calc_euclidean(anchor, negative)
        losses = torch.relu(distance_positive - distance_negative + self.margin)

        return losses.mean()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
def init_weights(m):
    if type(m) == nn.Linear:
        torch.nn.init.xavier_uniform_(m.weight)
        m.bias.data.fill_(0.01)
        
embedding_dims = 64

model = Network(embedding_dims)
model.apply(init_weights)
model = model.to(device)

from torch.utils.data import DataLoader, Dataset
train_loader = DataLoader(training_features, batch_size=1000, shuffle=False)

optimizer = optim.Adam(model.parameters(), lr=0.0001)
criterion = nn.TripletMarginLoss(margin=0.5, p=2.0)
#criterion = TripletLoss(margin = 0.5)
```

```python
model.train()
loss_c = []
epochs = 1
for epoch in tqdm(range(epochs),disable=True):
    running_loss = []
    for step, images in enumerate(tqdm(train_loader, desc="Training", leave=True,disable=True)):
        #print(images[:,:,0].shape)
        anchor_img = images[:,:,0].to(device)
        positive_img = images[:,:,1].to(device)
        negative_img = images[:,:,2].to(device)
        #print(anchor_img.shape)
        #print(anchor_img)
        optimizer.zero_grad()
        anchor_out = model(anchor_img)
        positive_out = model(positive_img)
        negative_out = model(negative_img)
        #print(anchor_out.shape)
        loss = criterion(anchor_out, positive_out, negative_out)
        #print(loss)
        loss.backward()
        optimizer.step()
        loss_c.append(loss.item())
        running_loss.append(loss.cpu().detach().numpy())
        #print('Loss:',loss.item(),'Image:',step,end = '\r')
        print('\r Epoch: {}/{} - Loss: {:.4f}'.format(epoch+1, epochs, np.mean(running_loss)),end = '')
    #print('\r Epoch: {}/{} - Loss: {:.4f}'.format(epoch+1, epochs, np.mean(running_loss)),end = '')
```
